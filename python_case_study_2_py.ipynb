```python
import random
import time

# Accept limits
min_limit = float(input("Enter Minimum Temperature Limit: "))
max_limit = float(input("Enter Maximum Temperature Limit: "))

# Number of readings (you can change this)
num_readings = 10

print("\nTemperature Monitoring System Started...\n")

for i in range(num_readings):
    current_temp = random.uniform(-10, 50)

    print(f"Reading {i+1}")
    print(f"Current Temperature: {current_temp:.2f}°C")

    if current_temp < min_limit:
        print("ALERT: Temperature is BELOW minimum limit!\n")
    elif current_temp > max_limit:
        print("ALERT: Temperature is ABOVE maximum limit!\n")
    else:
        print("Temperature is within safe limits.\n")

    time.sleep(2)

print("Monitoring Stopped.")
```
